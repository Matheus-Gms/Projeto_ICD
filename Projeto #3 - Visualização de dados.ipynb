# Projeto #3 - Visualização de dados


```python
import pandas as pd
import seaborn as sns
sns.color_palette("Paired")
```

```python
brasileirao = pd.read_csv("dataset/brasileirao_2015_2019.csv")
brasileirao
```

## Gráfico de Barras e de Linhas
#### Dados de Pontos (PTS) dos times divididos por região

### Região Norte

```python
norte = brasileirao[brasileirao["REGIÃO"] == "Norte"]
times_norte = norte.sort_values(by="TIMES")[["TIMES","ANO", "PTS"]]
times_norte
```

```python
sns.lineplot(data=times_norte, x="ANO", y="PTS", hue="TIMES", marker="o")
```

```python
sns.barplot(data=times_norte, x="ANO", y="PTS", hue="TIMES")
```

## Gráfico de Dispersão 
#### Correlação entre os dados de Vitórias (V) e Cartões Vermelhos (CV) dos 5 anos (2015/2019)

```python
get_together = brasileirao.groupby("TIMES").sum()
dispersal = sns.scatterplot(data= get_together , x="CV", y="V")
dispersal.set(xlabel="Cartões Vermelhos ",ylabel="Vitórias")
```

#### Correlação entre os dados Vitórias (V) e Gols Marcados (GP) dos 5 anos (2015/2019)

```python
dispersal2 = sns.scatterplot(data= get_together ,x ="GP",y="V")
dispersal2.set(xlabel="Gols Marcados",ylabel="Vitórias")
```
